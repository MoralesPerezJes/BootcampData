# Objetivo del proyecto

**Proyecto de Evaluación — Ingeniería de Datos: CRM + Billing + Universidad**

## Propósito

Construir, de principio a fin, un pipeline de datos que transforme información cruda de múltiples sistemas de origen en información útil y accionable para el negocio.

El proyecto simula el trabajo de un ingeniero de datos en una organización que opera con datos académicos, comerciales y de facturación. El reto no es solo mover datos, sino tomar decisiones técnicas justificadas en cada etapa del proceso.

## Qué se espera demostrar

- Capacidad para abordar un problema de datos de extremo a extremo.
- Criterio profesional al modelar, limpiar, transformar y automatizar el flujo de datos.
- Trazabilidad: cada decisión relevante queda documentada y argumentada.
- Reproducibilidad: el proyecto puede levantarse y ejecutarse desde cero.
- Comunicación: los resultados deben ser comprensibles también para perfiles no técnicos.

No existe una única solución correcta. Se evalúa el razonamiento aplicado y la calidad de la ejecución.

## Entregables del proyecto

- Pipeline automatizado (orquestado con Airflow).
- Capas de datos Bronze, Silver y Gold implementadas.
- Scripts de ingesta y transformación (Python y SQL).
- Notebooks de análisis con el razonamiento visible.
- Documentación de decisiones en `docs/decisiones.md`.
- Conclusiones e insights orientados al negocio.

## Secuencia de notebooks

El trabajo se documentará en notebooks independientes, cada uno con un foco concreto:

1. Objetivo del proyecto (este notebook)
2. Descripción de las fuentes de datos
3. Data Profiling
4. Limpieza y transformaciones realizadas
5. Análisis de posibles relaciones entre dominios y justificación
6. Implementación del pipeline Bronze -> Silver -> Gold
7. Cálculo de KPIs sobre Gold (`07_calculo_kpis.ipynb`)
8. Generación de tablas analíticas y/o archivos intermedios
9. Visualizaciones técnicas
10. Conclusiones y mejoras futuras
