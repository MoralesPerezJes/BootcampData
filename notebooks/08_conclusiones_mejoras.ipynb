# Conclusiones y mejoras futuras

Cierre del proyecto: resumen de lo analizado en los notebooks anteriores y mejoras sugeridas para futuras ediciones del bootcamp.

## 1. Conclusiones

El proyecto cubrió el flujo completo de un pipeline medallion sobre tres dominios (university, billing, crm): 18 tablas CSV, 447 500 filas, entregadas como sistemas de origen independientes.

Exploración y calidad (notebooks 01–03). Se inventariaron las fuentes, se perfilaron tipos, nulos y duplicados, y se detectaron situaciones concretas de calidad (emails repetidos, fechas invertidas, totales incoherentes, facturas sin ítems, pagos que exceden el total, status incorrectos, pesos de evaluación incompletos). Cada caso se investigó con datos y se tomó una decisión explícita para Silver.

Limpieza y relaciones (notebooks 04–05). Silver aplicó las transformaciones acordadas y registró cada intervención en la tabla incidencias de Postgres. El análisis entre dominios confirmó un único vínculo real (customers.external_ref = students.student_id); CRM no comparte llaves con billing ni university, por lo que Gold se modeló como constelación con hechos separados por dominio y dim_fecha conformada.

Pipeline y consumo (notebooks 06–07). Se automatizó Bronze → Silver → Staging → Gold con Airflow (medallion_pipeline), scripts en src/ y SQL en sql/gold/. Gold expone dimensiones, hechos y 6 vistas KPI (MRR, churn, ARPU, cobranza, aprobación/GPA/deserción, conversión y pipeline CRM, actividad por hora). Las decisiones quedan documentadas en docs/decisiones.md.

En conjunto, el entregable cumple el objetivo del bootcamp: datos crudos transformados en un modelo analítico reproducible, con trazabilidad de cada decisión técnica.

## 2. Mejoras futuras

Una de las partes a mejorar es organizar de mejor manera las instrucciones que debemos seguir, ya que al tener dos enfoques que chocan es difícil tomar una decisión, especialmente cuando no tienes experiencia.

También sería bueno tener un poco de aprendizaje y consejos a seguir para el análisis y resolución de situaciones.
